# Розв'язок

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
data = pd.read_csv('Position_Salaries.csv')
data
```

```python
x = data.Level
y = data.Salary
```

```python
plt.scatter(x,y)
plt.xlabel('Level')
plt.ylabel('Salary')
```

```python
model = np.poly1d(np.polyfit(x, y, 3))

polyline = np.linspace(1, 10)
plt.scatter(x, y)
plt.xlabel('Level')
plt.ylabel('Salary')
plt.plot(polyline, model(polyline))
plt.show()
```

```python
print(model)
```

```python
y_pred = model(x)
rmse = np.sqrt(np.mean(np.square(y-y_pred)))
print(f'RMSE for Polynomial Regression =>: {rmse:.2f}')
```

#### Порівняння із лінійною регресією

```python
lin_reg = LinearRegression()
```

```python
x_l = np.array(x)
y_l = np.array(y)
lin_reg.fit(x_l.reshape(-1,1), y_l.reshape(-1,1))
```

```python
y_pred1 = lin_reg.predict(x_l.reshape(-1,1))
                    
plt.figure(figsize=(6, 5))
plt.scatter(x, y)
plt.plot(x, y_pred1, color='b')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
print('RMSE for Linear Regression =>', round(np.sqrt(mean_squared_error(y, y_pred1)), 2))
```

```python
 Модель поміноміальної регресії точніша ніж лінійна модель. Це підтверджується розрахованими показниками RMSE і наглядно показано на побудованих графіках.
```
